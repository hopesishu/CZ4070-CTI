```python
# It is necessary to use Tor for DNS resolution of Onion websites

def getaddrinfo(*args):
    return [(socket.AF_INET, socket.SOCK_STREAM, 6, '', (args[0], args[1]))]

socket.getaddrinfo = getaddrinfo
```

```python
# Using beautifulsoup to get the website content into a nice format

soup = BeautifulSoup(res.content, 'html.parser')
```

```python
# Getting all links out of the soup and deleting None's

links = [link.get('href') for link in soup.find_all('a', href=True)]
links = list(filter(None, links))
links
```

```python
# Saving all links into a list

webpage_links = [] 
for l in links:
    l = XINGteam_link + l
    webpage_links.append(l)
        
webpage_links = webpage_links[1:-1] # remvoving first and last element from list
webpage_links
```

```python
info_list = []
header_list = []

for link in webpage_links:
    res = requests.get(link)
    soup = BeautifulSoup(res.content, 'html.parser')
    details_title = soup.find_all('dt')
    details_desc = soup.find_all('dd')

    for dt,dd in zip(details_title, details_desc):
        print(dt.text + ": " + dd.text)
    
    print("--------")
```

```python
info_list = []

for link in webpage_links:
    res = requests.get(link)
    soup = BeautifulSoup(res.content, 'html.parser')
    details_title = soup.find_all('dt')
    details_desc = soup.find_all('dd')
    
    info_row = []
    for dt in details_desc:
        info_row.append(dt.text.strip('\n'))
    info_list.append(info_row)
```

```python
import numpy as np
import pandas as pd

df = pd.DataFrame(info_list)
df.columns = ['Company Name', 'Total Revenue', 'Volume of Stolen Data', 'Published', 'Last Updated', 'Info']

df
```

```python
# pip install webdriver-manager
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

socket.socket = old_socket # reuse the original socket for selenium

s=Service(ChromeDriverManager().install())
print(s)

driver = webdriver.Chrome(service=s)
driver.maximize_window()
```

```python
# wikidpedia search
industry_list = []
number = 0

for company in df['Company Name']:
    driver.get('https://www.google.com')
    search = driver.find_element(By.NAME, 'q')
    search.send_keys(company + ' wikipedia')
    search.send_keys(Keys.RETURN) # hit return after you enter search text

    wiki_page = driver.find_element_by_tag_name('h3') # clicking the first search result
    wiki_page.click()

    try:
        infobox = driver.find_elements_by_class_name("infobox-label")
        infobox_data = driver.find_elements_by_class_name("infobox-data")

        index = 0
        for info in infobox:
            #print(info.text)
            if (info.text == "Industry"):
                industry_list.append(infobox_data[index].text)
                print(str(number) + ") " + company + ": " + infobox_data[index].text)
                break 
            index = index + 1
        else:
            industry_list.append(np.nan)
            print(str(number) + ") " + company + ": " + str(np.nan))
    except:
        print("cb")
    number = number + 1
    
driver.quit()
```

```python
# glassdoor search
industry_list = []
number = 0

for company in df['Company Name']: 
    driver.get('https://www.google.com')
    search = driver.find_element(By.NAME, 'q')
    search.send_keys(company + ' glassdoor "overview"')
    search.send_keys(Keys.RETURN)
    glassdoor_page = driver.find_element_by_tag_name('h3') # clicking the first search result
    glassdoor_page.click()
        
    try:
        pagesource = driver.page_source
        soup = BeautifulSoup(pagesource, 'html.parser')
        labels = soup.find_all('label')
        for label in labels:
            if (label.text == "Industry:"):
                industry = label.next_sibling.text
                industry_list.append(industry)
                print(str(number) + ") " + company + ": " + industry)
                break
        else:
            industry_list.append(np.nan)
            print(str(number) + ") " + company + ": " + str(np.nan))
    except:
        print("cb")
    number = number + 1
    
driver.quit()
```

```python
industry_df = pd.DataFrame(columns = ['Companies', 'Industries'])
industry_df['Companies'] = df['Company Name']
industry_df['Industries'] = industry_list

industry_df
```

```python
industry_df = industry_df.dropna()
industry_df = industry_df.reset_index(drop=True)

industry_df
```
