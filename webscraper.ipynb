```python
# Importing modules

import socks
import socket
import requests
from bs4 import BeautifulSoup
```

```python
# Configuring Socks to use Tor

from urllib.request import urlopen
socks.set_default_proxy(socks.SOCKS5, "localhost", 9150)
socket.socket = socks.socksocket
```

```python
# It is necessary to use Tor for DNS resolution of Onion websites

def getaddrinfo(*args):
    return [(socket.AF_INET, socket.SOCK_STREAM, 6, '', (args[0], args[1]))]

socket.getaddrinfo = getaddrinfo
```

```python
# Using requests package to read in the Hidden Wiki Onion Website on the Darknet

#res = requests.get("http://www.wikitjerrta4qgz4.onion/")
REvil_link = "http://dnpscnbaix6nkwvystl3yxglz7nteicqrou3t75tpcc5532cztc46qyd.onion"
res = requests.get(REvil_link)
```

```python
# Using beautifulsoup to get the website content into a nice format

soup = BeautifulSoup(res.content, 'html.parser')
```

```python
# Having a look at the Website content

#print(soup.prettify())
```

```python
# Checking the Websites title

soup.title
print(soup.title)
```

```python
# Getting all links out of the soup and deleting None's

links = [link.get('href') for link in soup.find_all('a')]
links = list(filter(None, links))
print(links)
```

```python
# Saving all links into a list

webpage_links = [] 
for l in links:
    if "page" in l:
        l = REvil_link + l
        webpage_links.append(l)
        
webpage_links = webpage_links[1:-1] # remvoving first and last element from list
webpage_links
```

```python
# Finding by class name

posts = soup.find_all(class_ = "blog-post-title")
for post in posts:
    title = post.find('a')
    print(title.text)
```

```python
master_list = []
for link in webpage_links:
    res = requests.get(link)
    soup = BeautifulSoup(res.content, 'html.parser')
    posts = soup.find_all(class_ = "blog-post-title")
    
    for post in posts:
        title = post.find('a')
        master_list.append(title.text)

master_list
# takes awhile to append all title to master_list
```

```python
import pandas as pd
df = pd.DataFrame(master_list, columns=['Companies'])
df
```
